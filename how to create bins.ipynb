```python
bin_labels_5 = ['Bronzfail', 'Silvergood', 'Goldverygood', 'Platinumexcellent', 'diamond']
df['quantile_ex_3'] = pd.qcut(df['TestScore'],
                              q=[0, .2, .4, .6, .8, 1],
                              labels=bin_labels_5)
df.head()
```
